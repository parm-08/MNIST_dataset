# MNIST (Modified National Institute of Standards and Technology) 
The MNIST dataset is like the "Hello World" of deep learning. 

It is a foundational dataset in machine learning, consisting of 70,000 grayscale images (28x28 pixels) of handwritten digits (0-9) used to train and test image classification models, serving as the "hello world" for computer vision due to its simplicity and effectiveness for benchmarking algorithms.

# WHAT WE WILL DO?
We have 28X28 pixels so 28*28=784 
so the input layer will have 784 neurons and we will create many hidden layers and output will tell us whether the digit is 0,1,2....,9 .
FINAL OUTPUT WILL HAVE 10 neurons(0-9 digits).

You’re working with the MNIST handwritten digits dataset and doing some basic setup and visualization before training a neural network.

```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)= tf.keras.datasets.mnist.load_data()  #here we are downloading the MNIST dataset with keras only.
```

It gives you:

x_train → training images

y_train → labels (correct answers) for training images

x_test → test images

y_test → labels for test images
